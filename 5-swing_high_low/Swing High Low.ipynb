```python
import pandas as pd
import numpy as np
%matplotlib inline
from pylab import mpl, plt
import seaborn as sns
from tqdm import notebook
#from matplotlib.finance import candlestick_ohlc
#from datamodule import data
```

```python
#get data
url = '../data/cleaned/currency/AUDCAD.pro_M15_201901020000_202107302345.csv'
df = pd.read_csv(url)
df.set_index('Time', inplace = True)
```

```python
#get swing high and low

#add swing high and low columns to df
df['swHigh'] = np.nan
df['swLow'] = np.nan
```

```python
#To get the position of the highest valueby rutrning a value n which will be minused from i
def getswingHigh(highlist):
    indexH = len(highlist)-highlist.index(max(highlist))+1
    return (max(highlist),indexH)

def getswingLow(lowlist):
    indexL = len(lowlist)-lowlist.index(min(lowlist))+1
    return (min(lowlist), indexL)
```

```python
curr_sHigh = df.High[1]
curr_sLow = df.Low[1]

highlist = []
lowlist =  []

sHighBreak = False
sLowBreak = False


for i in notebook.tnrange(1,len(df)):
    
    highlist.append(df.High[i])
    lowlist.append(df.Low[i])
    
    #break of swing high
    #change the conditional statement 
    if df.Close[i] > curr_sHigh:
        #set sHighBreak == True
        sHighBreak = True
        
    if df.Close[i] < curr_sLow:
        #set sLowBeak = True
        sLowBreak = True
        
    if(sHighBreak == True):
        # ensure no swing are filled
        # this will later be overidden
        df.iloc[i,4] = np.nan
        df.iloc[i,5] = np.nan
    if(sLowBreak ==  True):
        # ensure no swing are filled
        # this will later be overidden
        df.iloc[i,4] = np.nan
        df.iloc[i,5] = np.nan
        
    #look for swing set up
    if (sHighBreak == True and df.Low[i-1] > df.Close[i]):
        
        #create swing find fuction and pass values of i,close and low
        # fuction will search until a set up to find swing point
        curr_sHigh,indexH = getswingHigh(highlist)
        curr_sLow, indexL = getswingLow(lowlist)
        
        df.iloc[i,4] = curr_sHigh 
        df.iloc[i,5] = curr_sLow
                    
        sHighBreak = False
        sLowBreak = False

        # clear high/low list
        highlist = []
        lowlist =  []
            
            
    #look for swing set up
    if sLowBreak == True and df.High[i-1] < df.Close[i]:   
        #create swing find fuction and pass values of i,close and low
        # fuction will search until a set up to find swing point
        curr_sLow,indexL  = getswingLow(lowlist)
        curr_sHigh,indexH  = getswingHigh(highlist)
        #print(indexL,indexH)

        df.iloc[i,5] = curr_sLow
        df.iloc[i,4] = curr_sHigh
        
        sHighBreak = False
        sLowBreak = False

        # clear high/low list
        highlist = []
        lowlist =  []
        
    else:
        df.iloc[i,4] = curr_sHigh
        df.iloc[i,5] = curr_sLow
        
```

```python
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
df[-1000:].plot(title='AUDCAD', figsize=(15,8));
```
