```python
import pandas as pd
import numpy as np
%matplotlib inline
from pylab import mpl, plt
import seaborn as sns
from tqdm import notebook
#from matplotlib.finance import candlestick_ohlc
#from datamodule import data
```

```python
#get data
url = '../data/cleaned/currency/AUDCAD.pro_M15_201901020000_202107302345.csv'
df = pd.read_csv(url)
df.set_index('Time', inplace = True)
```

```python
#get swing high and low

#add swing high and low columns to df
df['swHigh'] = np.nan
df['swLow'] = np.nan
```

```python

#To get the position of the highest valueby rutrning a value n which will be minused from i

def getswingHigh(highlist):
    indexH = len(highlist)-highlist.index(max(highlist))+1
    return (max(highlist),indexH)

def getswingLow(lowlist):
    indexL = len(lowlist)-lowlist.index(min(lowlist))+1
    return (min(lowlist), indexL)
```

```python
df = df[:1000]
```

```python
# set current swing high and swing low as first value(least rescent value)
# done at initiatialization
curr_H = df.High[1]
curr_L = df.Low[1]
prev_H = np.nan
prev_L = np.nan

highlist = []
lowlist =  []


for i in notebook.trange(1,len(df)):
    
    highlist.append(df.High[i])
    lowlist.append(df.Low[i])
    
    # break of swing high
    # change the conditional statement 
    if df.Close[i] > curr_H:
        # set curr_H to prev_H
        prev_H = curr_H
        
    if prev_H == curr_H:
        # fill swing High with nan
        #df.iloc[i,4] = np.nan
        #df.iloc[i,5] = np.nan
        #look for swing set up
        if df.Low[i-1] > df.Close[i]:
            #create swing find fuction and pass values of i,close and low
            # fuction will search until a set up to find swing point
            curr_H,indexH = getswingHigh(highlist)
            curr_L,indexL  = getswingLow(lowlist)
            # clear high/low list
            highlist = []
            # clear high/low list
            lowlist =  [] 
            
            df.iloc[i,4] = curr_H 
            df.iloc[i,5] = curr_L
        
    if df.Close[i] < curr_L:
        # set curr_L to prev_L     
        prev_L = curr_L    
    if prev_L == curr_L:
        # fill swing low with nan
        #df.iloc[i,4] = np.nan
        #df.iloc[i,5] = np.nan
        #look for swing set up
        if df.High[i-1] < df.Close[i]:  
            #create swing find fuction and pass values of i,close and low
            # fuction will search until a set up to find swing point
            curr_L,indexL  = getswingLow(lowlist)
            curr_H,indexH = getswingHigh(highlist)
            # clear high/low list
            lowlist =  []   
            # clear high/low list
            highlist =  []
            
            df.iloc[i,4] = curr_H 
            df.iloc[i,5] = curr_L
            
        
    else:
        #set df['swHigh'] to current swing high
        df.iloc[i,4] = curr_H 
        df.iloc[i,5] = curr_L    
```

```python
plt.style.use('seaborn')
mpl.rcParams['savefig.dpi'] = 300
mpl.rcParams['font.family'] = 'serif'
df.plot(title='AUDCAD', figsize=(15,8));
```
